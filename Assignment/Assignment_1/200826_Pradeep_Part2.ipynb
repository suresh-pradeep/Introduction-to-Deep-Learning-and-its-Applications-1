The below block of code finds the mean of the computable features grouped by city. The floor was not used because some entries in the dataset had empty entries for the floor.

```python
df = pd.read_csv("House_prediction.csv")
city_group_mean=df.groupby(['city']).mean()
fig, ax = plt.subplots(9,1,figsize=(15,30))
ax[0].bar(city_group_mean.index,city_group_mean['area'])
ax[0].set_title("Mean Area")
ax[1].bar(city_group_mean.index,city_group_mean['rooms'])
ax[1].set_title("Mean number of Rooms")
ax[2].bar(city_group_mean.index,city_group_mean['bathroom'])
ax[2].set_title("Mean number of Bathrooms")
ax[3].bar(city_group_mean.index,city_group_mean['parking spaces'])
ax[3].set_title("Mean number of Parking Spaces")
ax[4].bar(city_group_mean.index,city_group_mean['hoa (R$)'])
ax[4].set_title("Mean hoa (R$)")
ax[5].bar(city_group_mean.index,city_group_mean['rent amount (R$)'])
ax[5].set_title("Mean rent amount (R$)")
ax[6].bar(city_group_mean.index,city_group_mean['property tax (R$)'])
ax[6].set_title("Mean property tax (R$)")
ax[7].bar(city_group_mean.index,city_group_mean['fire insurance (R$)'])
ax[7].set_title("Mean fire insurance (R$)")
ax[8].bar(city_group_mean.index,city_group_mean['total (R$)'])
ax[8].set_title("Mean total (R$)")
for ax1 in ax.flat:
    ax1.label_outer()
plt.show()
```

```python
#Plots for hoa
plt.scatter(df['rooms'].to_numpy(),df['hoa (R$)'].to_numpy())
plt.title('Rooms vs HOA')
plt.ylim(0,10000)
plt.show()
import seaborn as sns
fig, ax = plt.subplots(figsize=(8,6))
sns.scatterplot(x='area', y='hoa (R$)', hue='animal', data=df)
plt.title('Area vs HOA for different animal acceptance')
plt.xlim(0,1500)
plt.ylim(0,10000)
plt.show()
fig, ax = plt.subplots(figsize=(8,6))
sns.scatterplot(x='area', y='hoa (R$)', hue='furniture', data=df)
plt.title('Area vs HOA for different furniture status')
plt.xlim(0,1500)
plt.ylim(0,10000)
plt.show()
sns.scatterplot(x='area', y='hoa (R$)', hue='city', data=df)
plt.title('Area vs HOA for different cities')
plt.xlim(0,1500)
plt.ylim(0,10000)
plt.show()
plt.scatter(df['city'].to_numpy(),df['hoa (R$)'].to_numpy())
plt.title('City vs HOA')
plt.ylim(0,10000)
plt.show()

#Plots for property tax
plt.scatter(df['rooms'].to_numpy(),df['property tax (R$)'].to_numpy())
plt.title('Number of rooms vs Property tax')
plt.ylim(0,5000)
plt.show()
plt.scatter(df['city'].to_numpy(),df['property tax (R$)'].to_numpy())
plt.title('City vs Property tax')
plt.ylim(0,5000)
plt.show()
plt.scatter(df['area'].to_numpy(),df['property tax (R$)'].to_numpy())
plt.title('Area vs Property tax')
plt.xlim(0,1200)
plt.ylim(0,5000)
plt.show()

#Plots for fire insurance
plt.scatter(df['city'].to_numpy(),df['fire insurance (R$)'].to_numpy())
plt.title('City vs Fire insurance')
plt.ylim(0,600)
plt.show()
plt.scatter(df['furniture'].to_numpy(),df['fire insurance (R$)'].to_numpy())
plt.title('Furniture status vs Fire insurance')
plt.ylim(0,600)
plt.show()
plt.scatter(df['rent amount (R$)'].to_numpy(),df['fire insurance (R$)'].to_numpy())
plt.title('Rent amount vs Fire insurance')
plt.xlim(0,15000)
plt.ylim(0,600)
plt.show()
```

The overall trends are noticed as:
* The hoa, property tax and fire insurance vary by city
* The fire insurance shows a strong dependence with rent amount and it increases continously
* The property tax changes with the number of rooms and is maximum when the number of rooms is around 4 or 5
* The HOA also varies with the number of rooms and is maximum when the number of rooms is less than or equal to 5. The HOA also is higher among houses which allow animals
